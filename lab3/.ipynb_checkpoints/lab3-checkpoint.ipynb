```python
import pandas as pd
import matplotlib.pyplot as plt
import helpers 
import numpy

# Таблица данных
data = [
    {"t": 1, "tx": 14, "ty": 222},
    {"t": 2, "tx": 16, "ty": 241},
    {"t": 3, "tx": 17, "ty": 243},
    {"t": 4, "tx": 19, "ty": 285},
    {"t": 5, "tx": 18, "ty": 253},
    {"t": 6, "tx": 15, "ty": 247},
    {"t": 7, "tx": 14, "ty": 246},
    {"t": 8, "tx": 18, "ty": 276},
    {"t": 9, "tx": 17, "ty": 261},
    {"t": 10, "tx": 15, "ty": 254},
    {"t": 11, "tx": 13, "ty": 229},
    {"t": 12, "tx": 16, "ty": 249},
    {"t": 13, "tx": 17, "ty": 252},
    {"t": 14, "tx": 19, "ty": 279},
    {"t": 15, "tx": 18, "ty": 262},
    {"t": 16, "tx": 20, "ty": 275},
    {"t": 17, "tx": 12, "ty": 211},
    {"t": 18, "tx": 13, "ty": 220},
    {"t": 19, "tx": 12, "ty": 218},
    {"t": 20, "tx": 14, "ty": 232},
]

df = pd.DataFrame(data, columns=["t", "tx", "ty"])

# # print dataset (table)
df
```

```python
# Повторное применение преобразований и вывод расширенной таблицы
x_bar = df["tx"].mean()
y_bar = df["ty"].mean()

df_ext2 = df.copy()
df_ext2["tx^2"] = df_ext2["tx"] ** 2
df_ext2["(tx-x̄)^2"] = (df_ext2["tx"] - x_bar) ** 2
df_ext2["(ty-ȳ)^2"] = (df_ext2["ty"] - y_bar) ** 2
df_ext2["(tx-x̄)(ty-ȳ)"] = (df_ext2["tx"] - x_bar) * (df_ext2["ty"] - y_bar)

# print dataset (table)
df_ext2
```

```python
# Расчёт коэффициентов по формулам:
# a1^ = sum (tx - x̄)(ty - ȳ) / sum (tx - x̄)^2
# a0^ = ȳ - a1^ * x̄
x = df["tx"].to_numpy(dtype=float)
y = df["ty"].to_numpy(dtype=float)
T = len(df)

x_bar = x.mean()
y_bar = y.mean()

Sxx = numpy.sum((x - x_bar) ** 2)
Sxy = numpy.sum((x - x_bar) * (y - y_bar))

a1_hat = Sxy / Sxx
a0_hat = y_bar - a1_hat * x_bar

print(f"T = {T}")
print(f"x̄ = {x_bar:.3f}, ȳ = {y_bar:.3f}")
print(f"Sxx = {Sxx:.3f}, Sxy = {Sxy:.3f}")
print(f"a1^ = {a1_hat:.2f}")
print(f"a0^ = {a0_hat:.2f}")
```

```python
# Прогнозные значения, ошибки и вклад в оценку дисперсии
# x = df["tx"].to_numpy(dtype=float)
# y = df["ty"].to_numpy(dtype=float)
# T = len(df)

# # Коэффициенты модели (на случай, если предыдущая ячейка не запускалась)
# x_bar = x.mean()
# y_bar = y.mean()
# Sxx = numpy.sum((x - x_bar) ** 2)
# Sxy = numpy.sum((x - x_bar) * (y - y_bar))
# a1_hat = Sxy / Sxx
# a0_hat = y_bar - a1_hat * x_bar

# Столбцы модели
y_hat = a0_hat + a1_hat * df["tx"]
e = df["ty"] - y_hat
e2 = e ** 2
st2 = e2 / (T - 2)

model_table = pd.DataFrame({
    "t": df["t"],
    "tx": df["tx"],
    "ty": df["ty"],
    "ŷ": y_hat.round(3),
    "e": e.round(3),
    "e^2": e2.round(3),
    "s_t^2": st2.round(3),
})

# Строки сумма и среднее
sum_row = model_table.drop(columns=["t"]).sum(numeric_only=True)
mean_row = model_table.drop(columns=["t"]).mean(numeric_only=True)
sum_row["t"] = "сумма"
mean_row["t"] = "среднее"
summary = pd.DataFrame([sum_row, mean_row])[model_table.columns]

with pd.option_context('display.float_format', '{:.3f}'.format):
    display(model_table)
    display(summary)
```
